```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
from prophet import Prophet
import joblib
import os
```

```python
# Create a directory to save models
if not os.path.exists('saved_models'):
    os.makedirs('saved_models')

# Load synthetic data
df = pd.read_csv('synthetic_developer_data.csv')
```

```python
# Add time-based features
df['hour'] = pd.to_datetime(df['timestamp']).dt.hour
df['day_of_week'] = pd.to_datetime(df['timestamp']).dt.dayofweek  # Monday=0, Sunday=6
df['date'] = pd.to_datetime(df['timestamp']).dt.date

# Define metrics to analyze
metrics = ['typing_rhythm', 'tab_switches', 'errors', 'debugging', 'active_file_duration', 'idle_time']
```

```python
# Calculate a composite "focus score" (weighted average of metrics)
weights = {
    'typing_rhythm': 0.3,
    'active_file_duration': 0.3,
    'debugging': 0.2,
    'errors': -0.1,
    'idle_time': -0.1,
    'tab_switches': -0.1
}
df['focus_score'] = sum(df[metric] * weight for metric, weight in weights.items())

# Prepare features
features = ['hour', 'day_of_week']

# Train models and calculate relative scores
for metric in metrics:
    print(f"\nTraining model for {metric}...")

    # Prepare target
    y = df[metric]

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(df[features], y, test_size=0.2, random_state=42)

    # Train Random Forest model
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Evaluate model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Model Mean Squared Error for {metric}: {mse:.2f}")

    # Save the model
    model_filename = f'saved_models/{metric}_model.pkl'
    joblib.dump(model, model_filename)
    print(f"Model saved to {model_filename}")

    # Add predicted value to DataFrame
    df[f'predicted_{metric}'] = model.predict(df[features])

    # Calculate relative score
    df[f'relative_{metric}'] = df[metric] - df[f'predicted_{metric}']

    # Add interpretation
    df[f'performance_{metric}'] = np.where(
        df[f'relative_{metric}'] > 0, 'Better than expected',
        np.where(df[f'relative_{metric}'] < 0, 'Worse than expected', 'As expected')
    )
```

```python
# Time-Series Forecasting with Prophet
prophet_df = df[['timestamp', 'typing_rhythm']].rename(columns={'timestamp': 'ds', 'typing_rhythm': 'y'})
prophet_model = Prophet()
prophet_model.fit(prophet_df)

# Save the Prophet model
prophet_model_filename = 'saved_models/prophet_model.pkl'
joblib.dump(prophet_model, prophet_model_filename)
print(f"Prophet model saved to {prophet_model_filename}")

# Make predictions
future = prophet_model.make_future_dataframe(periods=24, freq='H')
forecast = prophet_model.predict(future)

# Plot forecast
prophet_model.plot(forecast)
plt.title('Typing Rhythm Forecast')
plt.show()

# Clustering with K-Means
X = df[['typing_rhythm', 'tab_switches', 'errors', 'debugging', 'active_file_duration', 'idle_time']]
kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(X)
```

```python
# Save the K-Means model
kmeans_model_filename = 'saved_models/kmeans_model.pkl'
joblib.dump(kmeans, kmeans_model_filename)
print(f"K-Means model saved to {kmeans_model_filename}")

# Visualize clusters
plt.figure(figsize=(10, 6))
for cluster in range(3):
    cluster_data = df[df['cluster'] == cluster]
    plt.scatter(cluster_data['hour'], cluster_data['focus_score'], label=f'Cluster {cluster}')
plt.title('Coding Session Clusters')
plt.xlabel('Hour of Day')
plt.ylabel('Focus Score')
plt.legend()
plt.grid()
plt.show()
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from prophet import Prophet
import joblib
import os

# Load saved models
def load_models():
    models = {}
    metrics = ['typing_rhythm', 'tab_switches', 'errors', 'debugging', 'active_file_duration', 'idle_time']
    
    for metric in metrics:
        model_filename = f'saved_models/{metric}_model.pkl'
        if not os.path.exists(model_filename):
            raise FileNotFoundError(f"Model file not found: {model_filename}")
        models[metric] = joblib.load(model_filename)
    
    # Load Prophet model
    prophet_model_filename = 'saved_models/prophet_model.pkl'
    if not os.path.exists(prophet_model_filename):
        raise FileNotFoundError(f"Prophet model file not found: {prophet_model_filename}")
    models['prophet'] = joblib.load(prophet_model_filename)
    
    # Load K-Means model
    kmeans_model_filename = 'saved_models/kmeans_model.pkl'
    if not os.path.exists(kmeans_model_filename):
        raise FileNotFoundError(f"K-Means model file not found: {kmeans_model_filename}")
    models['kmeans'] = joblib.load(kmeans_model_filename)
    
    return models

# Prepare new data for inference
def prepare_new_data(new_data):
    # Add time-based features
    new_data['hour'] = pd.to_datetime(new_data['timestamp']).dt.hour
    new_data['day_of_week'] = pd.to_datetime(new_data['timestamp']).dt.dayofweek  # Monday=0, Sunday=6
    return new_data

# Make predictions using the trained models
def make_predictions(models, new_data):
    # Prepare features
    features = ['hour', 'day_of_week']
    
    # Predict using Random Forest models
    for metric in models.keys():
        if metric not in ['prophet', 'kmeans']:
            new_data[f'predicted_{metric}'] = models[metric].predict(new_data[features])
            new_data[f'relative_{metric}'] = new_data[metric] - new_data[f'predicted_{metric}']
            new_data[f'performance_{metric}'] = np.where(
                new_data[f'relative_{metric}'] > 0, 'Better than expected',
                np.where(new_data[f'relative_{metric}'] < 0, 'Worse than expected', 'As expected')
            )
    
    # Predict using Prophet model
    prophet_df = new_data[['timestamp', 'typing_rhythm']].rename(columns={'timestamp': 'ds', 'typing_rhythm': 'y'})
    forecast = models['prophet'].predict(prophet_df)
    new_data['prophet_forecast'] = forecast['yhat']
    
    # Predict using K-Means model
    X = new_data[['typing_rhythm', 'tab_switches', 'errors', 'debugging', 'active_file_duration', 'idle_time']]
    new_data['cluster'] = models['kmeans'].predict(X)
    
    return new_data

# Generate actionable insights
def generate_insights(new_data):
    insights = []
    
    # Best and worst hours for productivity
    best_hour = new_data.groupby('hour')['typing_rhythm'].mean().idxmax()
    worst_hour = new_data.groupby('hour')['typing_rhythm'].mean().idxmin()
    insights.append(f"Best hour for productivity: {best_hour}:00")
    insights.append(f"Worst hour for productivity: {worst_hour}:00")
    
    # Debugging time analysis
    avg_debugging_time = new_data['debugging'].mean()
    if avg_debugging_time > new_data['debugging'].quantile(0.75):
        insights.append("High debugging time detected. Consider improving code quality or using better debugging tools.")
    
    # Idle time analysis
    avg_idle_time = new_data['idle_time'].mean()
    if avg_idle_time > new_data['idle_time'].quantile(0.75):
        insights.append("High idle time detected. Minimize distractions and focus on tasks.")
    
    # Typing rhythm analysis
    avg_typing_rhythm = new_data['typing_rhythm'].mean()
    if avg_typing_rhythm < new_data['typing_rhythm'].quantile(0.25):
        insights.append("Low typing rhythm detected. Consider taking breaks to avoid fatigue.")
    
    # Cluster analysis
    cluster_summary = new_data.groupby('cluster').agg({
        'typing_rhythm': 'mean',
        'debugging': 'mean',
        'idle_time': 'mean'
    }).reset_index()
    insights.append("Cluster Summary:")
    insights.append(cluster_summary.to_string(index=False))
    
    return insights

# Visualize predictions
def visualize_predictions(new_data):
    # Plot Prophet forecast
    plt.figure(figsize=(12, 6))
    plt.plot(new_data['timestamp'], new_data['typing_rhythm'], label='Actual Typing Rhythm')
    plt.plot(new_data['timestamp'], new_data['prophet_forecast'], label='Prophet Forecast', linestyle='--')
    plt.title('Typing Rhythm Forecast')
    plt.xlabel('Timestamp')
    plt.ylabel('Typing Rhythm')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Plot clusters
    plt.figure(figsize=(10, 6))
    for cluster in range(3):
        cluster_data = new_data[new_data['cluster'] == cluster]
        plt.scatter(cluster_data['hour'], cluster_data['typing_rhythm'], label=f'Cluster {cluster}')
    plt.title('Coding Session Clusters')
    plt.xlabel('Hour of Day')
    plt.ylabel('Typing Rhythm')
    plt.legend()
    plt.grid()
    plt.show()

# Main function
def main():
    try:
        # Load models
        models = load_models()
        
        # Load new data
        new_data = pd.read_csv('new_developer_data.csv')  # Replace with your new data file
        new_data = prepare_new_data(new_data)
        
        # Make predictions
        new_data = make_predictions(models, new_data)
        
        # Generate actionable insights
        insights = generate_insights(new_data)
        print("\nActionable Insights:")
        for insight in insights:
            print(f"- {insight}")
        
        # Visualize predictions
        visualize_predictions(new_data)
        
        # Save predictions to a CSV file
        new_data.to_csv('predictions.csv', index=False)
        print("\nPredictions saved to predictions.csv")
        
    except FileNotFoundError as e:
        print(f"Error: {e}")
        print("Please ensure all model files are saved in the 'saved_models' directory.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from prophet import Prophet
import joblib
import os

# Load saved models
def load_models():
    models = {}
    metrics = ['typing_rhythm', 'tab_switches', 'errors', 'debugging', 'active_file_duration', 'idle_time']
    
    for metric in metrics:
        model_filename = f'saved_models/{metric}_model.pkl'
        if not os.path.exists(model_filename):
            raise FileNotFoundError(f"Model file not found: {model_filename}")
        models[metric] = joblib.load(model_filename)
    
    # Load Prophet model
    prophet_model_filename = 'saved_models/prophet_model.pkl'
    if not os.path.exists(prophet_model_filename):
        raise FileNotFoundError(f"Prophet model file not found: {prophet_model_filename}")
    models['prophet'] = joblib.load(prophet_model_filename)
    
    # Load K-Means model
    kmeans_model_filename = 'saved_models/kmeans_model.pkl'
    if not os.path.exists(kmeans_model_filename):
        raise FileNotFoundError(f"K-Means model file not found: {kmeans_model_filename}")
    models['kmeans'] = joblib.load(kmeans_model_filename)
    
    return models

# Prepare new data for inference
def prepare_new_data(new_data):
    # Add time-based features
    new_data['hour'] = pd.to_datetime(new_data['timestamp']).dt.hour
    new_data['day_of_week'] = pd.to_datetime(new_data['timestamp']).dt.dayofweek  # Monday=0, Sunday=6
    return new_data

# Make predictions using the trained models
def make_predictions(models, new_data):
    # Prepare features
    features = ['hour', 'day_of_week']
    
    # Predict using Random Forest models
    for metric in models.keys():
        if metric not in ['prophet', 'kmeans']:
            new_data[f'predicted_{metric}'] = models[metric].predict(new_data[features])
            new_data[f'relative_{metric}'] = new_data[metric] - new_data[f'predicted_{metric}']
            new_data[f'performance_{metric}'] = np.where(
                new_data[f'relative_{metric}'] > 0, 'Better than expected',
                np.where(new_data[f'relative_{metric}'] < 0, 'Worse than expected', 'As expected')
            )
    
    # Predict using Prophet model
    prophet_df = new_data[['timestamp', 'typing_rhythm']].rename(columns={'timestamp': 'ds', 'typing_rhythm': 'y'})
    forecast = models['prophet'].predict(prophet_df)
    new_data['prophet_forecast'] = forecast['yhat']
    
    # Predict using K-Means model
    X = new_data[['typing_rhythm', 'tab_switches', 'errors', 'debugging', 'active_file_duration', 'idle_time']]
    new_data['cluster'] = models['kmeans'].predict(X)
    
    return new_data

# Generate fully data-specific actionable insights
def generate_insights(new_data):
    insights = []
    
    # Best and worst hours for productivity
    best_hour = new_data.groupby('hour')['typing_rhythm'].mean().idxmax()
    worst_hour = new_data.groupby('hour')['typing_rhythm'].mean().idxmin()
    insights.append(f"Your most productive hour is {best_hour}:00 (average typing rhythm: {new_data[new_data['hour'] == best_hour]['typing_rhythm'].mean():.2f}).")
    insights.append(f"Your least productive hour is {worst_hour}:00 (average typing rhythm: {new_data[new_data['hour'] == worst_hour]['typing_rhythm'].mean():.2f}).")
    insights.append(f"Recommendation: Schedule your most challenging tasks between {best_hour}:00 and {best_hour + 2}:00.")
    
    # Debugging time analysis
    avg_debugging_time = new_data['debugging'].mean()
    if avg_debugging_time > new_data['debugging'].quantile(0.75):
        insights.append(f"Your average debugging time is {avg_debugging_time:.2f} minutes, which is higher than the 75th percentile ({new_data['debugging'].quantile(0.75):.2f} minutes).")
        insights.append("Recommendation: Focus on writing cleaner code and use debugging tools like PyCharm or VS Code.")
    
    # Idle time analysis
    avg_idle_time = new_data['idle_time'].mean()
    if avg_idle_time > new_data['idle_time'].quantile(0.75):
        insights.append(f"Your average idle time is {avg_idle_time:.2f} minutes, which is higher than the 75th percentile ({new_data['idle_time'].quantile(0.75):.2f} minutes).")
        insights.append("Recommendation: Minimize distractions by turning off notifications and using focus apps like Focus@Will.")
    
    # Typing rhythm analysis
    avg_typing_rhythm = new_data['typing_rhythm'].mean()
    if avg_typing_rhythm < new_data['typing_rhythm'].quantile(0.25):
        insights.append(f"Your average typing rhythm is {avg_typing_rhythm:.2f}, which is lower than the 25th percentile ({new_data['typing_rhythm'].quantile(0.25):.2f}).")
        insights.append("Recommendation: Take regular breaks to avoid fatigue. Try the Pomodoro Technique (25 minutes of work, 5 minutes of break).")
    
    # Break scheduling
    insights.append(f"Recommendation: Take a 5-minute break every hour to recharge. Your average idle time is {avg_idle_time:.2f} minutes, which suggests you need more frequent breaks.")
    
    # Task prioritization
    insights.append(f"Recommendation: Break large tasks into smaller, manageable chunks. Your average active file duration is {new_data['active_file_duration'].mean():.2f} minutes, which suggests shorter, focused sessions work better for you.")
    
    # Energy management
    insights.append(f"Recommendation: Maintain energy levels by staying hydrated and eating healthy snacks during work sessions. Your typing rhythm drops significantly after {worst_hour}:00.")
    
    # Environment optimization
    insights.append(f"Recommendation: Optimize your workspace by reducing clutter and using noise-canceling headphones. Your idle time spikes during {worst_hour}:00, which may indicate distractions.")
    
    # Cluster analysis
    cluster_summary = new_data.groupby('cluster').agg({
        'typing_rhythm': 'mean',
        'debugging': 'mean',
        'idle_time': 'mean'
    }).reset_index()
    insights.append("Cluster Summary:")
    insights.append(cluster_summary.to_string(index=False))
    
    # Task-specific recommendations
    task_summary = new_data.groupby('task_type').agg({
        'typing_rhythm': 'mean',
        'debugging': 'mean',
        'idle_time': 'mean'
    }).reset_index()
    insights.append("Task-Specific Summary:")
    insights.append(task_summary.to_string(index=False))
    
    # Focus score analysis
    weights = {
        'typing_rhythm': 0.3,
        'active_file_duration': 0.3,
        'debugging': 0.2,
        'errors': -0.1,
        'idle_time': -0.1,
        'tab_switches': -0.1
    }
    new_data['focus_score'] = sum(new_data[metric] * weight for metric, weight in weights.items())
    avg_focus_score = new_data['focus_score'].mean()
    insights.append(f"Your average focus score is {avg_focus_score:.2f}.")
    insights.append("Recommendation: Focus on improving metrics with negative weights (e.g., reduce idle time and errors).")
    
    # Error rate analysis
    avg_errors = new_data['errors'].mean()
    if avg_errors > new_data['errors'].quantile(0.75):
        insights.append(f"Your average error rate is {avg_errors:.2f}, which is higher than the 75th percentile ({new_data['errors'].quantile(0.75):.2f}).")
        insights.append("Recommendation: Double-check your code and use automated testing tools to reduce errors.")
    
    # Tab switch analysis
    avg_tab_switches = new_data['tab_switches'].mean()
    if avg_tab_switches > new_data['tab_switches'].quantile(0.75):
        insights.append(f"Your average tab switches per hour is {avg_tab_switches:.2f}, which is higher than the 75th percentile ({new_data['tab_switches'].quantile(0.75):.2f}).")
        insights.append("Recommendation: Minimize context switching by focusing on one task at a time.")
    
    # Daily and weekly trends
    daily_trends = new_data.groupby('day_of_week')['typing_rhythm'].mean()
    insights.append("Daily Trends:")
    insights.append(daily_trends.to_string())
    
    # Anomaly detection
    anomalies = new_data[(new_data['idle_time'] > new_data['idle_time'].quantile(0.95)) | (new_data['typing_rhythm'] < new_data['typing_rhythm'].quantile(0.05))]
    insights.append("Anomalies Detected:")
    insights.append(anomalies[['timestamp', 'typing_rhythm', 'idle_time']].to_string(index=False))
    
    return insights

# Visualize predictions
def visualize_predictions(new_data):
    # Plot Prophet forecast
    plt.figure(figsize=(12, 6))
    plt.plot(new_data['timestamp'], new_data['typing_rhythm'], label='Actual Typing Rhythm')
    plt.plot(new_data['timestamp'], new_data['prophet_forecast'], label='Prophet Forecast', linestyle='--')
    plt.title('Typing Rhythm Forecast')
    plt.xlabel('Timestamp')
    plt.ylabel('Typing Rhythm')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Plot clusters
    plt.figure(figsize=(10, 6))
    for cluster in range(3):
        cluster_data = new_data[new_data['cluster'] == cluster]
        plt.scatter(cluster_data['hour'], cluster_data['typing_rhythm'], label=f'Cluster {cluster}')
    plt.title('Coding Session Clusters')
    plt.xlabel('Hour of Day')
    plt.ylabel('Typing Rhythm')
    plt.legend()
    plt.grid()
    plt.show()

# Main function
def main():
    try:
        # Load models
        models = load_models()
        
        # Load new data
        new_data = pd.read_csv('new_developer_data.csv')  # Replace with your new data file
        new_data = prepare_new_data(new_data)
        
        # Make predictions
        new_data = make_predictions(models, new_data)
        
        # Generate actionable insights
        insights = generate_insights(new_data)
        print("\nActionable Insights:")
        for insight in insights:
            print(f"- {insight}")
        
        # Visualize predictions
        visualize_predictions(new_data)
        
        # Save predictions to a CSV file
        new_data.to_csv('predictions.csv', index=False)
        print("\nPredictions saved to predictions.csv")
        
    except FileNotFoundError as e:
        print(f"Error: {e}")
        print("Please ensure all model files are saved in the 'saved_models' directory.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```
